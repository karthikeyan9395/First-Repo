```python
import pandas as pd
```

```python
data = pd.read_excel("D:\Analytics\Modern Trade\Analytics Transfer\Spar.XLSX",sheet_name = "Sheet1")
```

```python
data.columns
```

```python
import datetime as dt 
data['Date'] = '01' + '-' + data['Month'].str[0:3]+ '-' + data['Year'].map(str) 
data['Date'] = data['Date'].apply(lambda x: dt.datetime.strptime(x,'%d-%b-%Y'))
```

```python
new_data = data[data['P99.Item Description']== "MEERA HAIR WASH POWDER REFILL 80GM"]
# new_data = data
```

```python
new_data.sample(3)
```

```python
new_data_1 = new_data.groupby(["Date","P99.Item Description"]).agg(Min = ("SL.Net Amount","min"), 
                                          Max = ("SL.Net Amount","max"),
                                          Average = ("SL.Net Amount","mean"),
                                          Total = ("SL.Net Amount","sum")
                                                  ).reset_index()
```
